# Matplotlib Activities 
This notebook contains activities to practice various types of plots and visualizations using Matplotlib.

## Activity 1: Basic Line Plot
### Objective: Create a simple line plot to visualize data.
#### Instructions:
1. Create a list of x-values (`[0, 1, 2, 3, 4, 5]`) and a list of corresponding y-values (`[0, 1, 4, 9, 16, 25]`).
2. Use Matplotlib to plot these values as a line graph.
3. Label the axes as `x` and `y`.
4. Give the plot a title: `Basic Line Plot`.


```python
import matplotlib.pyplot as plt

# X and Y values
x_values = [0, 1, 2, 3, 4, 5]
y_values = [0, 1, 4, 9, 16, 25]

# Create line plot
plt.plot(x_values, y_values, marker='o', linestyle='-', color='b')

# Labels and title
plt.xlabel('x')
plt.ylabel('y')
plt.title('Basic Line Plot')

# Show plot
plt.show()
```

## Activity 2: Bar Plot
### Objective: Create a bar plot to visualize categorical data.
#### Instructions:
1. Create a list of categories (`['Category A', 'Category B', 'Category C', 'Category D']`) and corresponding values (`[4, 7, 3, 6]`).
2. Use Matplotlib to create a bar plot.
3. Add labels for the x and y axes.
4. Add a title to the plot: `Bar Plot Example`.

```python
import matplotlib.pyplot as plt

# Categories and values
categories = ['Category A', 'Category B', 'Category C', 'Category D']
values = [4, 7, 3, 6]

# Create bar plot
plt.bar(categories, values, color='blue')

# Labels and title
plt.xlabel('Categories')
plt.ylabel('Values')
plt.title('Bar Plot Example')

# Show plot
plt.show()
```

## Activity 3: Scatter Plot
### Objective: Create a scatter plot to visualize the relationship between two variables.
#### Instructions:
1. Create two lists of random values for x and y (e.g., `x = [1, 2, 3, 4, 5]`, `y = [5, 4, 3, 2, 1]`).
2. Use Matplotlib to plot a scatter plot.
3. Add labels for the axes and a title: `Scatter Plot Example`.

```python
import matplotlib.pyplot as plt
import random

try:
    # Generate random values for x and y
    x = [random.randint(1, 10) for _ in range(10)]
    y = [random.randint(1, 10) for _ in range(10)]

    # Ensure x and y have the same length
    if len(x) != len(y):
        raise ValueError("x and y lists must have the same length.")

    # Create scatter plot
    plt.scatter(x, y, color='blue', marker='o')

    # Annotate each point
    for i, (xi, yi) in enumerate(zip(x, y)):
        plt.text(xi, yi, f'({xi},{yi})', fontsize=9, ha='right')

    # Add labels and title
    plt.xlabel("X-axis")
    plt.ylabel("Y-axis")
    plt.title("Scatter Plot Example")

    # Show plot
    plt.show()

except ImportError as e:
    print(f"Error: {e}. Ensure required libraries are installed.")
except ValueError as e:
    print(f"Error: {e}")
except Exception as e:
    print(f"An unexpected error occurred: {e}")

```

## Activity 4: Histogram
### Objective: Create a histogram to visualize the distribution of a dataset.
#### Instructions:
1. Create a list of random values (e.g., `data = [1, 2, 2, 3, 4, 5, 5, 5, 6, 7]`).
2. Use Matplotlib to create a histogram with 5 bins.
3. Label the axes and give the plot a title: `Histogram Example`.

```python
import matplotlib.pyplot as plt
import numpy as np

# Generate random dataset
data = np.random.randint(1, 10, 20)

# Error handling for empty data
if len(data) == 0:
    print("Error: The dataset is empty. Please provide valid data.")
else:
    # Create histogram with 5 bins
    plt.hist(data, bins=5, edgecolor='black')

    # Labels and title
    plt.xlabel('Value')
    plt.ylabel('Frequency')
    plt.title('Histogram Example')

    # Show plot
    plt.show()
```

## Activity 5: Pie Chart
### Objective: Create a pie chart to visualize the proportions of different categories.
#### Instructions:
1. Create a list of categories (`['Apple', 'Banana', 'Cherry', 'Date']`) and their corresponding values (`[10, 15, 7, 5]`).
2. Use Matplotlib to create a pie chart.
3. Add a title to the chart: `Fruit Pie Chart`.

```python
import matplotlib.pyplot as plt

# Categories and values
categories = ['Apple', 'Banana', 'Cherry', 'Date']
values = [10, 15, 7, 5]

# Create pie chart
plt.pie(values, labels=categories, autopct='%1.1f%%', startangle=140)

# Title
plt.title('Fruit Pie Chart')

# Show plot
plt.show()

```

## Activity 6: Multiple Subplots
### Objective: Create multiple subplots to visualize different types of plots in one figure.
#### Instructions:
1. Create a figure with 2 rows and 2 columns of subplots.
2. In the first subplot, create a line plot.
3. In the second subplot, create a bar plot.
4. In the third subplot, create a scatter plot.
5. In the fourth subplot, create a pie chart.

```python
try:
    import matplotlib.pyplot as plt
except ImportError as e:
    print(f"Error importing matplotlib: {e}")
    exit(1)

# Create a 2x2 subplot layout
fig, axes = plt.subplots(2, 2, figsize=(10, 8))

# 1. Line chart in the first subplot (top-left)
x_line = [1, 2, 3, 4, 5]
y_line = [1, 4, 9, 16, 25]

if len(x_line) != len(y_line):
    print("Error: Line chart data lists must have the same length.")
    exit(1)

axes[0, 0].plot(x_line, y_line, marker='o')
axes[0, 0].set_title('Line Chart')
axes[0, 0].set_xlabel('X-axis')
axes[0, 0].set_ylabel('Y-axis')

# 2. Bar chart in the second subplot (top-right)
categories = ['A', 'B', 'C']
values = [10, 5, 15]

if len(categories) != len(values):
    print("Error: Bar chart data lists must have the same length.")
    exit(1)

axes[0, 1].bar(categories, values, color=['red', 'green', 'blue'])
axes[0, 1].set_title('Bar Chart')
axes[0, 1].set_xlabel('Categories')
axes[0, 1].set_ylabel('Values')

# 3. Scatter plot in the third subplot (bottom-left)
x_scatter = [1, 2, 3, 4]
y_scatter = [2, 4, 6, 8]

if len(x_scatter) != len(y_scatter):
    print("Error: Scatter plot data lists must have the same length.")
    exit(1)

axes[1, 0].scatter(x_scatter, y_scatter, color='magenta')
axes[1, 0].set_title('Scatter Plot')
axes[1, 0].set_xlabel('X-axis')
axes[1, 0].set_ylabel('Y-axis')

# 4. Pie chart in the fourth subplot (bottom-right)
pie_labels = ['Python', 'Java', 'C++']
pie_sizes = [50, 30, 20]

if sum(pie_sizes) != 100:
    print("Error: The sum of the pie chart sizes must equal 100 for a valid pie chart.")
    exit(1)

axes[1, 1].pie(pie_sizes, labels=pie_labels, autopct='%1.1f%%')
axes[1, 1].set_title('Pie Chart')

# Adjust spacing between subplots
plt.tight_layout()

# Display the plots
plt.show()
```
