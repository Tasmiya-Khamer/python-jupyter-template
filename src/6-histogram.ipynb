### Exercise 6: Histogram Customization

```python
# TASK: Create a histogram for the following data: data = [1, 2, 2, 3, 3, 3, 4, 4, 4, 4, 5].
# Customize the histogram with a title, labels for the x-axis, and a specific color for the bars.
try:
    import matplotlib.pyplot as plt
except ImportError as e:
    print(f"Error importing matplotlib: {e}")
    exit(1)

# Data
data = [1, 2, 2, 3, 3, 3, 4, 4, 4, 4, 5]

# Validate that data is not empty
if not data:
    print("Error: Data list is empty. Cannot create a histogram.")
    exit(1)

# Create the histogram
plt.hist(data, bins=range(1, 7), color='skyblue', edgecolor='black')  # bins=range(1,7) creates bins for each integer

# Add a title and axis labels
plt.title('Histogram of Values')
plt.xlabel('Value')
plt.ylabel('Frequency')

# Display the plot
plt.show()
```
