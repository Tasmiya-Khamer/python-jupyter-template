### Exercise 5: Subplot Layout

```python
# TASK: Create a 2x2 subplot layout.
# Plot a line chart in the first subplot, a bar chart in the second, a scatter plot in the third, and a pie chart in the fourth.
import matplotlib.pyplot as plt

# Create a 2x2 subplot layout
fig, axes = plt.subplots(2, 2, figsize=(10, 8))

# 1. Line chart in the first subplot (top-left)
x_line = [1, 2, 3, 4, 5]
y_line = [1, 4, 9, 16, 25]
axes[0, 0].plot(x_line, y_line, marker='o')
axes[0, 0].set_title('Line Chart')
axes[0, 0].set_xlabel('X-axis')
axes[0, 0].set_ylabel('Y-axis')

# 2. Bar chart in the second subplot (top-right)
categories = ['A', 'B', 'C']
values = [10, 5, 15]
axes[0, 1].bar(categories, values, color=['red', 'green', 'blue'])
axes[0, 1].set_title('Bar Chart')
axes[0, 1].set_xlabel('Categories')
axes[0, 1].set_ylabel('Values')

# 3. Scatter plot in the third subplot (bottom-left)
x_scatter = [1, 2, 3, 4]
y_scatter = [2, 4, 6, 8]
axes[1, 0].scatter(x_scatter, y_scatter, color='magenta')
axes[1, 0].set_title('Scatter Plot')
axes[1, 0].set_xlabel('X-axis')
axes[1, 0].set_ylabel('Y-axis')

# 4. Pie chart in the fourth subplot (bottom-right)
pie_labels = ['Python', 'Java', 'C++']
pie_sizes = [50, 30, 20]
axes[1, 1].pie(pie_sizes, labels=pie_labels, autopct='%1.1f%%')
axes[1, 1].set_title('Pie Chart')

# Adjust spacing between subplots
plt.tight_layout()

# Display the plots
plt.show()
```
